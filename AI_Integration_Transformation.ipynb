```python
import pandas as pd
import numpy as np
import warnings
```

```python
query = "SELECT * FROM AI_Adoption"  # Change to your table name
with warnings.catch_warnings():
    warnings.simplefilter("ignore", UserWarning)
    df = pd.read_sql(query, cnxn)

df.head()
```

### Which industry has the lowest adoption_rate?


```python
min_rate = df['adoption_rate'].min()
lowest_industries = df[df['adoption_rate'] == min_rate][['industry', 'adoption_rate']]
print(lowest_industries)
```

### What is the median adoption_rate across all countries?


```python
median_rate = df['adoption_rate'].median()
print( median_rate)
```

```python
median_countries = df[df['adoption_rate'] == median_rate]['country']
print("Countries with the median adoption rate:")
print(median_countries.to_list())
```

### Which countryindustry pair has the highest daily_active_users?


```python
max_dau = df['daily_active_users'].max()

# Filter the dataframe to get the country-industry pair(s) with that value
top_pair = df[df['daily_active_users'] == max_dau][['country', 'industry', 'daily_active_users']]
print(top_pair)
```

### Create a new column UserCategory = "High Usage" if daily_active_users > 5000 else "Low Usage".


```python

df['UserCategory'] = np.where(df['daily_active_users'] > 5000, 'High Usage', 'Low Usage')
df
```

### Display the first 10 rows

```python
df[['country', 'industry', 'daily_active_users', 'UserCategory']].head(10)
```

### Show the top 5 countries with the lowest adoption_rate in 2024.


```python

df_2024 = df[df['year'] == 2024]

lowest_5 = df_2024.sort_values('adoption_rate', ascending=True).head(5)

lowest_5[['country', 'adoption_rate']]

```

```python
df_2024 = df[df['year'] == 2024]

# Get the index of the row with minimum adoption_rate per country
idx = df_2024.groupby('country')['adoption_rate'].idxmin()

# Select those rows from df_2024
unique_countries = df_2024.loc[idx]

lowest_5_unique = unique_countries.sort_values('adoption_rate').head(5)

print(lowest_5_unique[['country', 'adoption_rate']])
```

### Which company_size has the highest mean daily_active_users?


```python
mean_dau_by_size = df.groupby('company_size')['daily_active_users'].mean()
max_value = mean_dau_by_size.max()

# Get all company sizes tied for the highest mean DAU
top_company_size = mean_dau_by_size[mean_dau_by_size == max_value]
print(top_company_size)
```

```python
type(mean_dau_by_size)
```

```python
max_value = mean_dau_by_size.max()

# Get the company size with max mean Dau (first match)
highest_company_size_label = mean_dau_by_size.idxmax()
print(f"Single company size with highest mean DAU: {highest_company_size_label}")
print(f"Max mean DAU value: {max_value}")
```

```python
min_value = mean_dau_by_size.min()

# Get all company sizes tied for the lowestest mean DAU
lowest_company_size = mean_dau_by_size[mean_dau_by_size == min_value]
print(lowest_company_size)
```

```python
min_value = mean_dau_by_size.min()

# Get the company size with min mean Dau (first match)
lowest_company_size_label = mean_dau_by_size.idxmin()
print(f"Company size with lowest mean DAU: {lowest_company_size_label}")
print(f"Lowest mean DAU value: {min_value}")
```

### Get summary statistics (mean, min, max) of adoption_rate grouped by AI tool

```python
summary_stats = df.groupby('ai_tool')['adoption_rate'].agg(['mean', 'min', 'max'])
print(summary_stats)
```

### From the DataFrame, select the first 10 rows where UserCategory is High Usage (daily_active_users > 5000).


```python
high_usage_first10 = df[df['UserCategory'] == 'High Usage'].head(10)
print(high_usage_first10)
```

### Create a Series of average adoption_rate by company_size using indexing, then display the top 2 company sizes with the highest values.


```python
avg_adoption_by_size = df.groupby('company_size')['adoption_rate'].mean()

top2_sizes = avg_adoption_by_size.sort_values(ascending=False).head(2)
print(top2_sizes)
```
